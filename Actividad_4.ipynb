## Ejercicio 4

Implementa una función llamada ejercicio4 que recibiendo el dataframe con los datos del Titanic, devuelva en una tupla el porcentaje de hombres y mujeres que viajaban en el Titanic, redondeados al segundo decimal.

```python
def ejercicio4(df):
    
    all_genders = df[["Sex"]]
    all_genders = all_genders.count()#numero total de tripulantes

    #reemplazamos las asignaciones de female y male 
    df["Sex_short"] = df["Sex"].replace({"male": "M", "female": "F"})

    #tripulantes hombres
    males = df[df["Sex_short"].str.contains("M")]
    males = males["Sex_short"]
    count_males = males.count()

    #tripulantes mujeres
    females = df[df["Sex_short"].str.contains("F")]
    females = females["Sex_short"]
    count_females = females.count()

    #porcentaje de tripulantes hombres y mujeres
    male_porcentual = (count_males/all_genders)*100
    female_porcentual = (count_females/all_genders)*100
    
    male_porcentual = round(male_porcentual, 2)
    female_porcentual = round(female_porcentual, 2)

    return male_porcentual,female_porcentual
```

```python
ejercicio4(df)
```

## Ejercicio 5


Implementa una función llamada ejercicio5 que recibiendo el dataframe con los datos del Titanic, devuelva en una lista el número de pasajeros que viajaban en 1ª, 2ª y 3ª clase.

```python
def ejercicio5(df):
    #pasajeros de primera clase
    first_class = df[(df["Pclass"] == 1)]
    first_class = first_class["Pclass"]
    count_first = first_class.count()

    #pasajeros de segunda clase
    second_class = df[(df["Pclass"] == 2)]
    second_class = second_class["Pclass"]
    count_second = second_class.count()

    #pasajeros de tercera clase
    third_class = df[(df["Pclass"] == 3)]
    third_class = third_class["Pclass"]
    count_third = third_class.count()

    lista = [count_first,count_second,count_third]
    
    return lista
```

```python
ejercicio5(df)
```
